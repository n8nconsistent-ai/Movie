# 🎬 Movie Recommendation System

## 📌 Project Overview
This project builds a content-based movie recommendation system using movie genres.
It recommends similar movies using cosine similarity.

## 🛠 Tools Used
- Python
- Pandas
- Scikit-learn
- CountVectorizer
- Cosine Similarity

## 📂 1. Importing Required Libraries

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## 📊 2. Loading the Dataset

The dataset contains movieId, title and genres columns.

```python
df=pd.read_csv('data.csv')
```

## 🧹 3. Data Cleaning & Preprocessing

- Removed extra spaces from titles
- Converted genres into text format
- Replaced "|" with space

```python
df['title'] = df['title'].str.strip()
```

```python
df['title'] = df['title'].str.strip()

df['genres'] = df['genres'].astype(str)
df['genres'] = df['genres'].str.replace('|', ' ', regex=False)
```

```python
type(df['genres'].iloc[0])
```

## 🔢 4. Text Vectorization

CountVectorizer converts text data into numerical vectors.

```python
cv = CountVectorizer(stop_words='english')
vectors = cv.fit_transform(df['genres'])
```

## 📐 5. Calculating Cosine Similarity

Cosine similarity measures how similar two movies are.

```python
similarity = cosine_similarity(vectors)
```

```python
pickle.dump(df, open('movies.pkl','wb'))
pickle.dump(similarity, open("similarity.pkl", "wb"))
```

## 🎯 6. Recommendation Function

This function takes a movie name as input and returns top similar movies.

```python
movies = df['title'].values

def recommend(movie_name, top_n=5):
    movie_name = movie_name.lower()
    movie_list_lower = [movie.lower() for movie in movies]

    if movie_name not in movie_list_lower:
        return "Movie not found"

    index = movie_list_lower.index(movie_name)
    distances = list(enumerate(similarity[index]))
    distances = sorted(distances, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i in distances[1:top_n+1]:
        recommendations.append(movies[i[0]])

    return recommendations
```

## 🧪 7. Testing the Recommendation System

```python
recommend("Another Stakeout (1993)")
```
